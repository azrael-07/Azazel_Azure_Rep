```python
def co(x):
    return 5+np.power((x-4),2)
```

```python
def grad(x):
    return 2*(x-5)
```

```python
def graddesc(x,learn_rate=0.01,iterate=100,print_all=True):
    cost=[]
    for i in range(iterate):
        x=x-learn_rate*grad(x)
        if i%50==0 and print_all:
            print("the iteration is {} and curr val of x is {}".format(i,x))
        cost.append(co(x))
    return x,cost
```

```python
lr=0.01
n_itr=200
X=np.random.random()*10
X,cost=graddesc(X,lr,n_itr)
print("max is at (x,y)= ({},{})".format(X,co(X)))
```

```python
fig,ax=plt.subplots(figsize=(10, 8))
_=ax.plot(range(n_itr),cost,'b.')
```

```python
lr1=0.1
n_itr1=50
X1=np.random.random()*10
X11,cost1=graddesc(X1,lr1,n_itr1)
print("max is at (x,y)= ({},{})".format(X11,co(X1)))

fig,ax=plt.subplots(figsize=(10, 8))
_=ax.plot(range(n_itr1),cost1,'b.')
```

```python
learning_rates = [0.01,0.1,1,10]
iterations = 200
X_0 = np.random.random()
plt.figure(figsize=(20,5))
all_costs = []
for lr in learning_rates:
    X, costs = graddesc(X_0,lr,iterations, print_all=True)
    print("Minimum is at (x,y) = ({}, {})".format(X, co(X)))
    all_costs.append(costs)

ax.set_ylabel('Cost')
ax.set_xlabel('Iterations')
plt.xlim((0,200))
plt.ylim((0,50))
plt.scatter(range(iterations), all_costs[0], color='red')
plt.scatter(range(iterations), all_costs[1], color='blue')
plt.scatter(range(iterations), all_costs[2], color='green')
plt.scatter(range(iterations), all_costs[3], color='black')
plt.legend(['LR = 0.01', 'LR = 0.1', 'LR = 1', 'LR = 10'])
plt.show()
```
